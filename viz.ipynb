```python
from re import A
import altair as alt
import pandas as pd
from vega_datasets import data

alt.renderers.enable("mimetype")    

source = pd.read_csv(r"C:\Users\Acrylic\Desktop\Chatbot\Web scraper\Master File ImobiliareRoNou.csv")
chartMax = alt.data_transformers.disable_max_rows()
```

```python
from matplotlib.pyplot import title

alt.Chart(source).mark_point().encode(
    y=alt.Y('areaRaw:Q' ,axis=alt.Axis(title='Suprafta')),
    x=alt.X('price:Q', axis=alt.Axis( title='Pret')),
    size= 'rooms:Q',
    color= alt.Color('rooms:Q',scale = alt.Scale(range = ['FF3333','#9C27B0','#70235B','#1f9fb4'])),
    tooltip= [alt.Tooltip(field ='price', title = 'Pret', type ='quantitative'),
              alt.Tooltip(field ='floor', title = 'Etaj', type = 'quantitative'),
              alt.Tooltip(field ='eur/mp', title = 'Eur/mp', type = 'quantitative'),
              alt.Tooltip(field ='rooms', title = 'Nr_camere', type = 'quantitative'),
              alt.Tooltip(field ='areaRaw', title = 'Suprafata', type = 'quantitative'),
              alt.Tooltip(field = 'location', title = 'Locatie', type = 'nominal')
             ] 
).interactive() 
```

```python
chart =alt.Chart(source).mark_circle().encode(
    alt.X(alt.repeat("column") , type = 'quantitative'),
    alt.Y(alt.repeat("row"), type= 'quantitative'),
    alt.Color('type', type = 'nominal',
             scale = alt.Scale(range = ['#3F51B5','#9C27B0','#70235B','#1f9fb4','#FF5733'])),
  
    tooltip= [alt.Tooltip(field ='price', title = 'Pret', type ='quantitative'),
              alt.Tooltip(field ='floor', title = 'Etaj', type = 'quantitative'),
              alt.Tooltip(field ='eur/mp', title = 'Eur/mp', type = 'quantitative'),
              alt.Tooltip(field ='rooms', title = 'Nr_camere', type = 'quantitative'),
              alt.Tooltip(field ='areaRaw', title = 'Suprafata', type = 'quantitative'),
              alt.Tooltip(field = 'location', title = 'Locatie', type = 'nominal')
             ]
).properties(
   
    width  = 500,
    height = 250
).repeat(
        row = ['price','areaRaw','rooms'],
        column = ['rooms', 'areaRaw' , 'price']
).interactive()     

alt.vconcat(chart,title = 'Distibution of the locations according to their extincion date and their connections')

chart.save('chart.html')
```

```python
from pandas import interval_range


histArea = alt.Chart(source).mark_bar(clip=True).encode(
    y=alt.Y('count(area_rounded):Q' ,axis=alt.Axis(title='Nr de apartamente cu suprafata de ')),
    x=alt.X('area_rounded:Q', axis=alt.Axis( title='Suprafata'),scale=alt.Scale(domain=(20, 70))),
    tooltip = [alt.Tooltip ('count()', title = 'Nr de apartamente cu suprafata de')]
)

histPrice = alt.Chart(source).mark_bar(clip=True).encode(
    y=alt.Y('count(priceRounded):Q' ,axis=alt.Axis(title='Nr de apartamente cu pretul de ')),
    x=alt.X('area_rounded:Q', axis=alt.Axis( title='Pret'), scale=alt.Scale(domain=(30, 60))),
    tooltip = [alt.Tooltip ('count()', title = 'Nr de apartamente cu pretul de')]
)

histPriceRange = alt.Chart(source).mark_bar().encode(
    y=alt.Y('count(priceRange):Q' ,axis=alt.Axis(title='Nr de apartamente cu intervalul de pret de ')),
    x=alt.X('priceRange:N', axis=alt.Axis( title='Pret')),
    tooltip = [alt.Tooltip ('count()', title = 'Nr de apartamente cu intervalul de pret de')]
)

histAreaRange = alt.Chart(source).mark_bar().encode(
    y=alt.Y('count(area_range):Q' ,axis=alt.Axis(title='Nr de apartamente cu suprafta de ')),
    x=alt.X('area_range:N', axis=alt.Axis( title='Suprafta interval')),
    tooltip = [alt.Tooltip ('count()', title = 'Nr de apartamente cu suprafta de')]
)

chart = histArea & histPrice | histPriceRange & histAreaRange

```

```python
histLoc = alt.Chart(source).mark_bar().encode(
    y=alt.Y('count(approxLoc):Q' ,axis=alt.Axis(title='Nr de apartamente in locatia de ')),
    x=alt.X('approxLoc:N', axis=alt.Axis( title='Locatie')),
    tooltip = [alt.Tooltip ('count()', title = 'Nr de apartamente in locatia de'),
               alt.Tooltip('mean(price)', title = 'Medie pret')]
)

histRooms = alt.Chart(source).mark_bar().encode(
    y=alt.Y('count(rooms):Q' ,axis=alt.Axis(title='Nr de apartamente pt fiecare tip ')),
    x=alt.X('rooms:O', axis=alt.Axis( title='Camere')),
    tooltip = [alt.Tooltip ('count()', title = 'Nr de apartamente pt fiecare tip'),
               alt.Tooltip('mean(price)', title = 'Medie pret')]
)

histFloor = alt.Chart(source).mark_bar().encode(
    y=alt.Y('count(floor):Q' ,axis=alt.Axis(title='Nr de apartamente la etajul')),
    x=alt.X('floor:O', axis=alt.Axis( title='Etaj')),
    tooltip = [alt.Tooltip ('count()', title = 'Nr de apartamente la etajul'),
               alt.Tooltip('mean(price)', title = 'Medie pret')]
)
chart4 = histLoc | histRooms | histFloor
chart4
```
